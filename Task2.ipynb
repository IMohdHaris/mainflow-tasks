```python
# prompt: 1. Learn how to read CSV files and manipulate
# data frames using Pandas.
# 2. Perform simple data cleaning tasks such as
# handling missing values and removing
# duplicates.
# 3. Practice basic data manipulation operations
# like filtering, sorting, and grouping data.

import pandas as pd

# 1. Read CSV and create a DataFrame
df = pd.read_csv('/content/sample_data/california_housing_test.csv')
print(df)

# 2. Handle missing values
df.fillna(df.mean(), inplace=True)
print(df)

# Remove rows with any missing values
df.dropna(inplace=True)
print(df)

# Remove duplicate rows
df.drop_duplicates(inplace=True)
print(df)

# 3. Data manipulation
# Filter data
filtered_df = df[df['median_income'] > 10]
print(filtered_df)

# Sort data
sorted_df = df.sort_values('median_income')
print(sorted_df)

# Group data
grouped_df = df.groupby('median_income').mean()
print(grouped_df)
```
